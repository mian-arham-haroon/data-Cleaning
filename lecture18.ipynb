```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('loan.csv')
df
```

```python
df.isnull().sum()
```

```python
df['ApplicantIncome']= df['ApplicantIncome'].fillna(df['ApplicantIncome'].mean())
```

```python
sns.distplot(df['ApplicantIncome'], bins=20, kde=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
ss.fit(df[['ApplicantIncome']])
df['new_application']=pd.DataFrame(ss.transform(df[['ApplicantIncome']]),columns=['new_application'])
df['new_application']
df
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.distplot(df['new_application'], bins=20, kde=True)
plt.subplot(1,2,2)
sns.distplot(df['ApplicantIncome'], bins=20, kde=True)
plt.show()
```

```python
sns.boxplot(x=df['new_application'], palette='Set2', orient='h')
plt.title('Boxplot of new_application')
```
